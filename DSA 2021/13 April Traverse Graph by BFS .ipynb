```python
class Graph:
    def __init__(self):
        self.listofairposts = {}
    
    def addconnection(self, city, connection):
        if city in self.listofairposts.keys():
            self.listofairposts[city].append(connection)
        else:
            self.listofairposts[city] = [connection]
```

```python
airports = Graph()
```

```python
# Lots of info we have to store like in graph, so we have graph database 
airports.addconnection("Jai", "Mum")
airports.addconnection("Jai", "Del")
airports.addconnection("Jai", "Hyd")
airports.addconnection("Mum", "Jai")
airports.addconnection("Mum", "Del")
airports.addconnection("Mum", "Bang")
airports.addconnection("Del", "Jai")
airports.addconnection("Del", "Mum")
airports.addconnection("Del", "Bang")
airports.addconnection("Bang", "Goa")
airports.addconnection("Bang", "Mum")
airports.addconnection("Bang", "Del")
airports.addconnection("Hyd", "Goa")
airports.addconnection("Hyd", "Jai")
airports.addconnection("Goa", "Bang")
airports.addconnection("Goa", "Hyd")
```

```python
def bfs(firstvertex):
    # Mark and also contain traversal info
    visited = []
    queue = []
    
    # Enqueue
    queue.append(firstvertex)
    
    # Ask adjacents and store in queue
    adjacentcity = airports.listofairposts[firstvertex]
    for i in adjacentcity:
        queue.append(i)
    
    # After store firstvertex and their adjacents in queue, mark firstvertex(Append in visited)
    # Marking visited
    if firstvertex not in visited:
        visited.append(firstvertex)
    
    # Deque firstvertex after take info of their adjacents and marked
    queue.pop(0)
    print("Marked : ", visited)
    print("Current queue : ", queue)
    bfs(queue[0])
```

```python
bfs('Jai')
```

```python
def bfs_city(vertex):
    visited = [vertex]
    queue = [vertex]
    while queue:
        devertex = queue.pop(0)
        for adjacentvertex in airports.listofairposts[devertex]:
            if adjacentvertex not in visited:
                visited.append(adjacentvertex)
                queue.append(adjacentvertex)
    return visited
```
