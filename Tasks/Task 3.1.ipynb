Take 2 images and combine it to form single image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('task3.1(1).jpg')
img2 = cv2.imread('task3.1(2).jpg') 
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
print(img1.shape)
print(img2.shape)
```

```python
img1 = cv2.resize(img1 , (259 , 194))
```

```python
print(img1.shape)
print(img2.shape)
```

```python
showimg = np.hstack((img1 , img2))
```

```python
plt.imshow(showimg)
```

```python
showimg=cv2.cvtColor(showimg, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(showimg)
```
