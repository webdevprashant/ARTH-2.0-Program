Take 2 images, crop some part of both the images and swap them

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('task3.1(1).jpg')
img2 = cv2.imread('task3.1(2).jpg')
```

```python
print(img1.shape)
print(img2.shape)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
cropimg1 = img1[10:120 , 105:230]
plt.imshow(cropimg1)
```

```python
cropimg2 = img2[30:130 , 90:180]
plt.imshow(cropimg2)
```

```python
cropimg1= cv2.resize(cropimg1,(90,100))
cropimg2= cv2.resize(cropimg2,(125,110))
```

```python
plt.imshow(cropimg1)
```

```python
plt.imshow(cropimg2)
```

```python
img1[10:120 , 105:230] = cropimg2
img2[30:130 , 90:180] = cropimg1
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img1=cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
print(img1.shape)
print(img2.shape)
```

```python
img1 = cv2.resize(img1 , (259,194))
```

```python
newimg = np.hstack((img1,img2))
```

```python
plt.imshow(newimg)
```
